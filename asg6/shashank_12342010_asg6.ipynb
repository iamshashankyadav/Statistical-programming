```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
for n1 in [10,1000,10000,100000]:
    data=np.random.normal(20, 2, n1)
    mean=np.mean(data)
    std=np.std(data)
    plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
    
    # Overlay the normal PDF using the estimated mean and std
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mean, std)
    plt.plot(x, p, 'k', linewidth=2)
    
    # Labels and title
    plt.title(f"Histogram of Simulated Battery Life and PDF\nMean: {mean:.2f}, Std: {std:.2f},n1:{n1:}")
    plt.xlabel('Battery Life (hours)')
    plt.ylabel('Density')
    plt.show()
```

```python
def plot_estimators_distribution(means, std_devs,n1, true_mu=20, true_sigma=2):
    # Plot histograms for the estimated means and standard deviations
    plt.figure(figsize=(12, 6))
    print("n2=",len(means),"n1=",n1)
    # Plot for the means
    plt.subplot(1, 2, 1)
    plt.hist(means, bins=30, density=True, alpha=0.6, color='red')
    plt.axvline(true_mu, color='blue', linestyle='dashed', linewidth=2, label=f'True Mean: {true_mu}')
    plt.title('Estimated Mean Distribution for n')
    plt.xlabel('Estimated Mean')
    plt.ylabel('Density')
    
    # Plot for the standard deviations
    plt.subplot(1, 2, 2)
    plt.hist(std_devs, bins=30, density=True, alpha=0.6, color='green')
    plt.axvline(true_sigma, color='blue', linestyle='dashed', linewidth=2, label=f'True Std Dev: {true_sigma}')
    plt.title('Estimated Std Dev Distribution')
    plt.xlabel('Estimated Std Dev')
    plt.ylabel('Density')
    
    plt.tight_layout()
    plt.show()

n2_range=[100,1000]
for i in n2_range:
    
    for n1 in [100,1000]:
        total_mean=[]
        total_std=[]
        for n2 in range(i):
            data=np.random.normal(20, 2, n1)
            total_mean.append(np.mean(data))
            total_std.append(np.std(data))
        
        plot_estimators_distribution(total_mean, total_std, n1,true_mu=20, true_sigma=2)
```

```python
for n1 in  [10, 100, 1000,10000]:
    X = np.random.normal(50, 5, n1)
    # Simulate sensor noise  uniformly distributed between -1 and 1
    noise = np.random.uniform(-1, 1, n1)
    # Measured temperature Y = X + 
    Y = X + noise
    # MLE for the mean and std dev from the noisy data is just the sample mean and std of Y
    mean = np.mean(Y)
    sigma = np.std(Y)
 
    plt.hist(Y, bins=30, density=True, alpha=0.6, color='g')
    
    # Overlay the normal PDF using the estimated mean and std of X
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mean, sigma)
    plt.plot(x, p, 'k', linewidth=2)
    
    # Labels and title
    plt.title(f"Histogram of Noisy Measurements and PDF\nEstimated Mean: {mean:.2f}, Estimated Std: {sigma:.2f}")
    plt.xlabel('Measured Temperature (Y)')
    plt.ylabel('Density')
    plt.show()
```

```python
mu_3=0.1
sig_3=2
var_3=sig_3**2
deg_f=(2*var_3/(var_3-1))
print(f"Actual Mean: {mu_3}")
print(f"Actual Standard Deviation: {sig_3}")
print(f"Actual Variance: {var_3}")
print(f"Degree of Freedom: {deg_f}")
```

```python
from scipy.stats import t
for n in [100,1000,10000]:
  obs_3=np.random.standard_t(df=deg_f,size=n)+mu_3
  mu_3_calc=np.mean(obs_3)
  sd_3_calc=np.std(obs_3)
  print(f"Caluclated Mean:{mu_3_calc}")
  print(f"Calculated Standard Deviation:{sd_3_calc}")
  plt.title(f"Student T Distribution n= {n}")
  plt.xlabel("x")
  plt.ylabel("f(x)")
  plt.hist(obs_3,range=(-max(abs(min(obs_3)),abs(max(obs_3))),max(abs(min(obs_3)),abs(max(obs_3)))),density=True,bins=max(int(max(obs_3)-min(obs_3)),10))
  x=np.linspace(min(obs_3),max(obs_3),max(1000,n))
  pdf=t.pdf(x,df=deg_f,loc=mu_3_calc,scale=sd_3_calc)
  plt.plot(x,pdf)
  plt.show()
```

```python
for n in [100,1000,10000]:
  obs_3_n=np.random.standard_t(df=deg_f,size=n)+mu_3
  noise_3_n=np.random.uniform(-0.5,0.5,n)
  Y_3=[obs_3_n[i]+noise_3_n[i] for i in range(n)]
  mu_3_calc_n=np.mean(obs_3)
  sd_3_calc_n=np.std(obs_3)
  print(f"Caluclated Mean:{mu_3_calc_n}")
  print(f"Calculated Standard Deviation:{sd_3_calc_n}")
  plt.title(f"Student T Distribution n= {n}")
  plt.xlabel("x")
  plt.ylabel("f(x)")
  plt.hist(Y_3,range=(-max(abs(min(Y_3)),abs(max(Y_3))),max(abs(min(Y_3)),abs(max(Y_3)))),density=True,bins=max(int(max(Y_3)-min(Y_3)),10))
  x=np.linspace(min(Y_3),max(Y_3),max(1000,n))
  pdf=t.pdf(x,df=deg_f,loc=mu_3_calc_n,scale=sd_3_calc_n)
  plt.plot(x,pdf)
  plt.show()
```
