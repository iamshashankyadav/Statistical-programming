```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv("dataset_q1/data_1 - data_1.csv")
df2=pd.read_csv("dataset_q1/data_2 - data_2.csv")
df1=df1.T
df2=df2.T
```

```python
corr1=pd.DataFrame(df1).corr()  #
corr2=pd.DataFrame(df1).corr()
display(corr1.head())
corr2.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(-1, 1))
df1_normalized = pd.DataFrame(scaler.fit_transform(df1))
df2_normalized = pd.DataFrame(scaler.fit_transform(df2))
```

```python
corr1norm=pd.DataFrame(df1_normalized).corr()
corr2norm=pd.DataFrame(df2_normalized).corr()
display(corr1norm.head())
corr2norm.head()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=10)  # Reduce to 10 dimensions
df1_pca = pca.fit_transform(df1_normalized.T)
df2_pca = pca.fit_transform(df2_normalized.T)
display(df1_pca.shape)
# Compute correlation matrices again after PCA transformation
corr_matrix_1_pca = pd.DataFrame(df1_pca).corr()
corr_matrix_2_pca = pd.DataFrame(df2_pca).corr()

```

```python
j=0
for i in [corr1,corr1norm,corr_matrix_1_pca]:
    j+=1
    plt.figure(figsize=(10, 8))
    sns.heatmap(i, cmap="coolwarm", annot=False)
    if j==1:
        plt.title("Correlation Matrix - Data 1 (Original)")
    elif j==2:
        plt.title("Correlation Matrix - Data 1(normalized)")
    else:
        plt.title("Correlation Matrix - Data 1 (Pca applied)")
    plt.show()
```

```python
j=0
for i in [corr2,corr2norm,corr_matrix_2_pca]:
    j+=1
    plt.figure(figsize=(10, 8))
    sns.heatmap(i, cmap="coolwarm", annot=False)
    if j==1:
        plt.title("Correlation Matrix - Data 2 (Original)")
    elif j==2:
        plt.title("Correlation Matrix - Data 2(normalized)")
    else:
        plt.title("Correlation Matrix - Data 2 (Pca applied)")
    plt.show()
```

```python
from scipy.stats import chi2  
for n in [1,100,1000,100000]:   # for n in this range
    num=np.random.normal(0,1,n)   # take random numer from normal distribution
    mean=0
    variance=1
    x=(num-mean)**2    # do the transformation formula
    x=x/variance    
    
    sns.histplot(x, bins=30, kde=True, stat="density", alpha=0.5, color="red")  # now plot the converted values and thier kde plot to smoothen the graph
    
    plt.title(f"Histogram and KDE for Sample Size {n}")
    plt.xlabel("Value")
    plt.ylabel("Density")
    x_vals = np.linspace(0, max(x), 1000)
    y_vals = chi2.pdf(x_vals, df=1)

  
    plt.plot(x_vals, y_vals, color="blue", linestyle="--", linewidth=2, label="Chi-Square (df=1)")  # plot the theroticalpdf of chi square for the comparison

    plt.show()
    
```

```python

from scipy.stats import norm
data = pd.read_csv("dataset_q3/gaussian_dataset_with_noise.csv") 
for i in data.index:    # convert the strings value to float 
    data.loc[i,"Value"]=float(data.loc[i,"Value"])


mean_val = np.mean(data)   # Compute mean and variance
variance_val = np.var(data)   # Compute mean and variance 

std_dev = np.sqrt(variance_val) #compute the std variance
print("the mean and varience of the data is ",mean_val,"and",float(variance_val))
# Compute percentage of data points within 1σ, 2σ, and 3σ
within_1sigma = np.mean((data >= mean_val - std_dev) & (data <= mean_val + std_dev)) * 100    # calculating the data points lie t= sigma range and t=taking oercentage
within_2sigma = np.mean((data >= mean_val - 2 * std_dev) & (data <= mean_val + 2 * std_dev)) * 100   # same as above
within_3sigma = np.mean((data >= mean_val - 3 * std_dev) & (data <= mean_val + 3 * std_dev)) * 100
print('the mean of the data points within the the sigma 1 is',within_1sigma)
print('the mean of the data points within the the sigma 2 is',within_2sigma)
print('the mean of the data points within the the sigma 3 is',within_3sigma)
# Compute the CDF of the dataset

sorted_values = np.sort(data)
cdf_values = norm.cdf(sorted_values, loc=mean_val, scale=std_dev)   # have usedthe inv=buolt fn to calculate the cdf

# Probability of data points beyond 2σ (both tails)
prob_beyond_2sigma = 1 - (norm.cdf(mean_val + 2 * std_dev, loc=mean_val, scale=std_dev) - 
                          norm.cdf(mean_val - 2 * std_dev, loc=mean_val, scale=std_dev))
print("the probability of data points falling beyond 2σ",prob_beyond_2sigma[0])


plt.figure(figsize=(8, 6))
plt.plot(sorted_values, cdf_values,'o', label="CDF", color='b')
plt.xlabel("Data Values")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF)")
plt.grid()
plt.legend()
plt.show()
```
